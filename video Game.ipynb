```python
plt.figure(figsize=(6,4))
sns.set_context('talk')
sns.countplot(y=df["Genre"], order=df['Genre'].value_counts().index, palette='coolwarm')
plt.title('Top-Genre')
plt.xlabel('Genre-Counts')
plt.ylabel('Genres')
```

```python
#Top selling regions
import matplotlib.pyplot as plt
sum_sales=df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
plt.figure(figsize=(6,4))
sns.barplot(x=sum_sales.index, y=sum_sales.values, palette='viridis')
 #title
plt.xlabel('Regions')
plt.ylabel('TOTAL-SALES')
plt.show()
```

```python
#Popular platform 
#popularity for the platform  we will now use countplot 
plt.figure(figsize=(8,6))
plt.style.use('bmh')

sns.countplot(y=df['Platform'], order=df['Platform'].value_counts().index,
              palette='magma',width=0.9)
plt.title('Platform-Popularity')
plt.xlabel('Counts')
plt.ylabel('Platforms-Type')
plt.show()
```

```python
sale=df.groupby('Year')['Global_Sales'].sum().reset_index()
```

```python
# Global sales history over the Year
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1)
sns.lineplot(x='Year', y='Global_Sales', data=sale, marker='o')
plt.title('Global-Sales-History')
plt.show()
```

```python
top_publisher=df.groupby('Publisher')['Global_Sales'].sum().nlargest(10).reset_index()
```

```python
# Top Publishers by Global-sales.
plt.style.use('fivethirtyeight')
sns.barplot(x="Global_Sales", y="Publisher", data=top_publisher, palette='dark')
plt.title('Top-Publishers')
plt.xlabel('Global Sales')
plt.ylabel('Publishers')
plt.show()
```
